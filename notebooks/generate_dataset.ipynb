```python
import requests
from bs4 import BeautifulSoup
import os
from PIL import Image
import random

# Function to download images from Google Images


def download_images(query, label, max_images):
   # Create a directory to store the images
    base_path = "../data"
    if not os.path.exists(base_path):
        os.makedirs(base_path)
    if not os.path.exists(os.path.join(base_path, label)):
        os.makedirs(os.path.join(base_path, label))

    # Create the Google Images search URL
    url = "https://www.google.com/search?q={}&tbm=isch".format(query)
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    # Send a GET request to the URL
    r = requests.get(url, headers=headers)

    # Parse the HTML response using BeautifulSoup
    soup = BeautifulSoup(r.text, 'html.parser')

    # Find all image tags in the HTML response
    images = soup.find_all('img')

    # Download the images and save them to the directory
    count = 0
    for image in images:
        try:
            # Get the image source URL and send a GET request to it
            src = image['src']
            r = requests.get(src, headers=headers)

            # Save the image to the directory with the label as the filename prefix
            filename = "{}{}.jpg".format(label, count)
            with open(os.path.join(base_path, label, filename), 'wb') as f:
                f.write(r.content)

            count += 1
            if count == max_images:
                break

        except:
            continue


# Download images of people posing for pictures and label them as "pose"
download_images("people posing for pictures", "pose", 1000)

# Download images of people standing around and label them as "stand"
download_images("people standing around", "stand", 1000)
```

```python
# Display some random samples
pose_files = os.listdir(os.path.join("data", "pose"))
stand_files = os.listdir(os.path.join("data", "stand"))

print("Samples of people posing for pictures:")
for i in range(3):
    file = random.choice(pose_files)
    img = Image.open(os.path.join("data", "pose", file))
    display(img)

print("Samples of people standing around:")
for i in range(3):
    file = random.choice(stand_files)
    img = Image.open(os.path.join("data", "stand", file))
    display(img)
```
